```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definiera lite matriser och vektorer

matrix1 = np.array([[0,0],[0,0.16]])
matrix2 = np.array([[0.85,0.04],[-0.04,0.85]])
matrix3 = np.array([[0.2,-0.26],[0.23,0.22]])
matrix4 = np.array([[-0.15,0.28],[0.26,0.24]])
vector1 = np.array([[0],[1.6]])
vector2 = np.array([[0],[0.44]])
```

```python
# Skapa lite funktioner

def function1(column_vector: np.ndarray) -> np.ndarray:
    return matrix1@column_vector

def function2(column_vector: np.ndarray) -> np.ndarray:
    return matrix2@column_vector + vector1

def function3(column_vector: np.ndarray) -> np.ndarray:
    return matrix3@column_vector + vector2

def function4(column_vector: np.ndarray) -> np.ndarray:
    return matrix4@column_vector + vector2
```

```python
# Skapa en bild med hjälp av slumpen och våra funktioner
def get_image(n_iter=500000):
    coordinates = [np.array([[0.],[0.]])]
    for i in range(n_iter):
        k=np.random.randint(1,101)
        if k==1:
            coordinates.append(function1(coordinates[i]))
        elif k < 86:
            coordinates.append(function2(coordinates[i]))
        elif k < 93:
            coordinates.append(function3(coordinates[i]))
        else:
            coordinates.append(function4(coordinates[i]))
    x=np.array([x[0][0] for x in coordinates])
    y=np.array([x[1][0] for x in coordinates])
    return x,y

# Skapa och plotta en bild med hjälp av get_image ovan
def plot_image():
    x, y = get_image()
    plt.figure(figsize=(8,8),facecolor='black')
    plt.scatter(x,y,c='green',s=0.05)
    plt.gca().set_facecolor('black')
    plt.show()
```

```python
# Testa
plot_image()
```
